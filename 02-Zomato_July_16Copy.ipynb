```python
# Dependencies
import requests
import json
import pandas as pd

# Google developer API key
from config import gkey
```

```python
# variables
target_city = "305" # entity id for Denver
target_search = "55"
target_radius = 80 
# in meters, about 5 miles


# set up a parameters dictionary
params = {
    "apikey": gkey,
    "start": 0,
    "city_id": target_city,
    "radius": target_radius,
    "cuisines": target_search,
    "sort": "cost", "rating"
    "order": "desc",
    "count": 20
}

# base url
base_url = "https://developers.zomato.com/api/v2.1/search"

# run a request using our params dictionary
response = requests.get(base_url, params=params)
response.url
```

```python
# print the response url, avoid doing for public github repos in order to avoid exposing key
print(response.url)
# print('https://developers.zomato.com/api/v2.1/search?q=Denver&radius=80&cuisines=25')
```

```python
# convert response to json
places_data = response.json()

# Print the json (pretty printed)
print(json.dumps(places_data, indent=4, sort_keys=True))  
```

```python
resp = requests.get(base_url, params).json()
```

```python
resp
```

```python
def get_restaurant(resp, canonical_food_type): 
    rows = []
    for i in resp.get("restaurants"):
        q = i.get("restaurant").get("location").get("city")
        target_search = i.get("restaurant").get("cuisines")
        name = i.get("restaurant").get("name")
        review = i.get("restaurant").get("user_rating").get("aggregate_rating")
        price = i.get("restaurant").get("average_cost_for_two")
        rows.append({
            "Food_Type": target_search,
            "Ultimate_Food_Type": canonical_food_type,
            "Restaurant": name,
            "Rating": review,
            "City": q,
            "Price": price
            
        })
        
#  print(f'Food_Type:{target_search} \n Restaurant:{name} \n Rating:{review} \n City:{q}')
    return rows
```

```python
Den_Ital = pd.DataFrame.from_records(get_restaurant(resp, "Italian"))
Den_Ital
```
