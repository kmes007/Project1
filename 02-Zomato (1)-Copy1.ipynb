```python
# create a df with the list of cuisisines
df = pd.DataFrame.from_records([x.get('cuisine') for x in resp.get('cuisines')]).head()

# function to query the df based on a logical searhc input like "Ameri"
def get_ids(search_str):
    ids =  df[df['cuisine_name'].str.contains(search_str)].cuisine_id.tolist()
    stringids = list(map(str, ids))
    return ",".join(stringids)

get_ids('Chi')
```
